```python
import pandas as pd
```

```python
col = ['nid', "tipo_conciliacion",'expediente', 'empresa', 'nombre_empresa', 'fecha', 'estado',
         'nombre', 'conciliador']

df = pd.read_csv("raw/estados1trimestre.csv", names = col)
```

```python
ac = df[df.estado == 'Conciliación cerrada con acuerdo'].groupby('conciliador').estado.value_counts()
ac=pd.DataFrame(ac)
sa= df[df.estado == 'Conciliación cerrada sin acuerdo'].groupby('conciliador').estado.value_counts()
sa=pd.DataFrame(sa)
income= df[df.estado == 'Conciliación cerrada por ausencia injustificada de la empresa'].groupby('conciliador').estado.value_counts()
income=pd.DataFrame(income)
incomr = df[df.estado == 'El consumidor no asistió a la primera audiencia y no justificó su ausencia'].groupby('conciliador').estado.value_counts()
incomr=pd.DataFrame(incomr)
incomdos = df[df.estado == 'Ni el consumidor ni la empresa asistieron a la audiencia y no justificaron su ausencia'].groupby('conciliador').estado.value_counts()
incomdos=pd.DataFrame(incomdos)
asignado= df.conciliador.value_counts()
asignado=pd.DataFrame(asignado).reset_index()
asignado.columns = ['conciliador','asignadas']
```

```python
final0 = pd.merge(left = ac, right = sa,
                       how = "left", left_on = "conciliador", right_on = "conciliador")
final1 = pd.merge(left = final0, right = income,
                       how = "left", left_on = "conciliador", right_on = "conciliador")
final2 = pd.merge(left = final1, right = incomr,
                       how = "left", left_on = "conciliador", right_on = "conciliador")
final3 = pd.merge(left = final2, right = incomdos,
                       how = "left", left_on = "conciliador", right_on = "conciliador")

analisisconc = pd.merge(left = final3, right = asignado,
                       how = "inner", left_on = "conciliador", right_on = "conciliador")


analisisconc.columns = ['conciliador','con acuerdo', 'sin acuerdo', 'incomparecencia empresa', 'incomparecencia requirente', 
                  'incomparecencia ambos', 'asignadas']
```

```python
analisisconc.to_excel('tidy/analisisconci.xlsx')
```
